```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Number of Nodes = 500
n = 500
# Average degree of network = k
k = []
for i in range(10,51):
    k.append(i/10.0)
print("Average Degrees: ")
print(k)
p = []
for i in k:
    p.append(i/(n-1))
print("Probabilities: ")
print(p)
```

```python
G = []
for prob in p:
    G.append(nx.erdos_renyi_graph(n, prob))
```

```python
'''
graphs = []
for grp in G:
    graphs.append(list(nx.connected_component_subgraphs(grp)))
'''
g_max = []
for ccs in G:
    g_max.append(max(nx.connected_component_subgraphs(ccs), key=len))
#print g_max.number_of_nodes()
```

```python
Ng = []
for n_g in g_max:
    Ng.append(len(n_g))
```

```python
# x axis values 
X = k
# corresponding y axis values 
Y = []
for y in Ng:
    Y.append(y/(n*1.0))

# plotting the points  
plt.plot(X, Y)
  
# naming the x axis 
plt.xlabel('X - axis: <k>') 
# naming the y axis 
plt.ylabel('Y - axis: (Ng/N)') 
  
# giving a title to my graph 
plt.title('Evolution of Random Network') 
  
# function to show the plot 
plt.show() 
```

```python
nx.draw(G[0], with_labels=False,node_color = "b",edge_color = "green")
plt.savefig('figure1.png')
plt.show()
```
